```python
import pandas as pd
```

```python
df=pd.read_csv("data.csv")
df_region= pd.read_csv('noc.csv')
```

```python
df
```

```python
df=df[df['Season']=='Summer']
```

```python
df
```

```python
df_region
```

## hear we have to add region column for our observaltion

### we use merge operation df 

```python
df=df.merge(df_region,on='NOC',how='left')
```

```python
df.head()
```

## finding duplicates row and remove it

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace =True)
```

```python
df.duplicated().sum()
```

### Now we have to focus on medals

```python
df["Medal"].value_counts()
```

### we use one hot encoding 
- using get_dummies function

```python
pd.get_dummies(df['Medal'])
```

### concat this data frame with df

```python
df=pd.concat([df,pd.get_dummies(df['Medal'])],axis=1)
```

```python
df.head()
```

```python
df.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False)
```

```python
medal_tally=df.drop_duplicates(subset=['Team','NOC','Games','Year','City','Sport','Event','Medal'])
```

```python
medal_tally.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
```
